```python
sections = data.section_JSON_to_dict("scraping/classes.json")
print(len(sections))
student_dict = data.generate_students_weighted(sections, 3000)
for key in student_dict:
    student_dict[key].find_conflicts(sections)
```

```python

def monte_carlo_gs(student_dict: dict, section_dict: dict, iteration_limit: int = 500) -> int:
    best = float("inf")
    seeds = [i for i in range(1, iteration_limit + 2)]
    i = 0
    while i < iteration_limit:
        i += 1
        
        new_student_dict = copy.deepcopy(student_dict)
        new_section_dict = copy.deepcopy(section_dict)
        new_matching = gs.gale_shapley_match(new_student_dict, new_section_dict, shuffle = True, seed = seeds[i])
        
        rogues = test_stability.is_weakly_stable(new_matching[0], new_matching[1])[1]
        
        if rogues < best:
            best = rogues
            
    return int(best)
```

```python
monte_carlo_gs(student_dict, sections, 10)
```
